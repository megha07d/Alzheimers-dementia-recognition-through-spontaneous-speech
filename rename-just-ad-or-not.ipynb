```python
# all imports here

# file ops
import os

# for copying directories
import shutil

# zip download
from IPython.display import FileLink
```

```python
# final folder


dir_name = 'dementia-audio-classify'
path = '/kaggle/working/' + dir_name

# create
os.makedirs(path)

# check
if os.path.exists(path):
    print(f'{dir_name} successfully created')
else:
    print(f'{dir_name} not created, try again')
```

```python
# function 
# input - dir_path
# output - list of files/dirs in it


def list_contents(dir_path):
    count = 0
    for filename in os.listdir(dir_path):
        print(filename)
        count += 1
    if count%10 == 0:
        print(count)
```

```python
def count_files(dir_path):
    count = 0
    for filename in os.listdir(dir_path):
        print(filename)
        
           
    #vprint(count)
    
    
dir_path ='/kaggle/working/dementia-audio-classify'
count_files(dir_path)
```

```python
# function 
# input - directory of folders
# output - good named files in output folder


def simplify_dir(dir_path,ad):
    
    count = 0
    
    new_dir_path = '/kaggle/working/dementia-audio-classify'

    for folder_name in os.listdir(dir_path):
        # print(folder_name)

        folder_path = os.path.join(dir_path,folder_name)
        # print(folder_path)
        for filename in os.listdir(folder_path):
            # print(filename)
            parent_file_name = folder_name.split('-')[0]

            adrso_name = parent_file_name[:5]
            subj_num = parent_file_name[5:]

            name_1 = filename.split('.')[0]
            name_2 = name_1.split('_')[1]

            extension = os.path.splitext(filename)[1]
            
            if ad is True:
                addon = 'ad'
            else:
                addon = 'cn'
            new_file_name = adrso_name + '_' + subj_num + '_' + name_2 + '_' + addon + extension
            count += 1
            # print(new_file_name)



            # now rename
            new_file_path = os.path.join(new_dir_path,new_file_name)
            old_file_path = os.path.join(folder_path,filename)

    #         print(new_file_path)
    #         print(old_file_path)

            os.rename(old_file_path,new_file_path)
        # print(count)

```

```python
# create sample_AD
# first copy input dir to output working dir - coz kaggle allows read-only for input
source_dir_ad = '/kaggle/input/dementia-audio-data/up_sampled/ad'
destin_dir_ad = '/kaggle/working/sample_AD'
if os.path.exists(destin_dir_ad):
    shutil.rmtree(destin_dir_ad)
shutil.copytree(source_dir_ad,destin_dir_ad)



# create sample_CN
# first copy input dir to output working dir - coz kaggle allows read-only for input
source_dir_cn = '/kaggle/input/dementia-audio-data/up_sampled/cn'
destin_dir_cn = '/kaggle/working/sample_CN'
if os.path.exists(destin_dir_cn):
    shutil.rmtree(destin_dir_cn)
shutil.copytree(source_dir_cn,destin_dir_cn)
```

```python
# simplify AD samples

dir_path_ad = '/kaggle/working/sample_AD'
simplify_dir(dir_path_ad,True)



# simplify CN samples

dir_path_cn = '/kaggle/working/sample_CN'
simplify_dir(dir_path_cn,False)
```
